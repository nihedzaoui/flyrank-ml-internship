## 1. My lane as an ML task (type)

**Task Type:** Ranking

My lane focuses on ranking content based on its predicted usefulness and engagement. Instead of predicting a simple yes/no outcome, the model estimates which pieces of content should appear higher in search or recommendation results.

A ranking model is appropriate because users interact with multiple pieces of content, and the goal is to order them from most to least relevant.

```python
# No computation is required for this section.

task_type = "Ranking"
print("ML Task:", task_type)
```

## 2. Target or proxy

The model predicts a content quality score using observed user engagement.

Since "content quality" is not directly measurable, a proxy target is used. Examples include:

- Click-through rate (CTR)
- Average reading time
- Number of conversions
- User engagement score

These values come from observed user behavior rather than manually defined rules.

```python
import pandas as pd

df = pd.DataFrame({
    "content_id":[1,2,3,4,5],
    "title":[
        "AI trends 2026",
        "Machine Learning Guide",
        "Python Tutorial",
        "SEO Best Practices",
        "Data Science Roadmap"
    ],
    "impressions":[5000,6000,4000,5500,7000],
    "clicks":[250,420,160,330,560],
    "engagement_score":[0.5,0.7,0.4,0.6,0.8]
})

df.head()
```

## 3. Success metric

Primary metric:

NDCG (Normalized Discounted Cumulative Gain)

NDCG measures how well the model ranks the most relevant content near the top of the list.

A higher NDCG score indicates better ranking quality.

For an initial baseline, an NDCG above 0.80 would indicate good performance.

```python
print("Chosen Metric: NDCG")
```

## 4. The unit of analysis

One row represents one content item.

Each row contains features describing the content, such as:

- title

- engagement metrics

The dataframe below shows the unit of analysis used by the model.

```python
print(df.head())

print("\nShape:")
print(df.shape)

print("\nColumns:")
print(df.columns.tolist())
```

## 5. Why ML beats a fixed rule

A fixed rule such as

"If CTR is greater than 5%, rank first"

cannot capture the complexity of user behavior.

Content performance depends on many interacting factors, including:

- search intent
- keyword competition
- content quality
- freshness
- user engagement
- historical performance

Machine learning can learn these complex relationships from data and produce better rankings than manually designed rules.

```python
print("Machine learning captures complex patterns that fixed rules cannot.")
```
